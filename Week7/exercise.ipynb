```python
plt.scatter(df_mpg['displacement'], df_mpg['acceleration'])
plt.xlabel('displacement (cm)')
plt.ylabel('acceleration (cm)')
plt.grid()
```

```python
toyota = df_mpg[df_mpg.model_year == 78][df_mpg.car_name.str.startswith('to')]
others = df_mpg[df_mpg.model_year == 78][~df_mpg.car_name.str.startswith('to')]
others
toyota
# setosa_data = data[data.species == 'setosa']
# versicolor_data = data[data.species == 'versicolor']
# virginica_data = data[data.species == 'virginica']

fig, ax=plt.subplots(1,2,figsize=(13, 5))


ax[0].hist(toyota.cylinders, color='g', label = 'toyota')
ax[1].hist(others.cylinders, color='r', label = 'other cars')


ax[0].legend()
ax[1].legend()

ax[0].set_ylabel('cars')
ax[1].set_ylabel('cars')

ax[0].set_xlabel('cylinders')
ax[1].set_xlabel('cylinders')


plt.show()
```
